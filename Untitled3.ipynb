```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data.csv")
df.info()
```

```python
df.head(10)
```

**SEARCHING FOR ANY NULL DATA + *CLEANING***



```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Certification'])
```

```python
df['MetaScore'].fillna(df['MetaScore'].mean(), inplace=True)
df['Gross'].fillna(df['Gross'].median(), inplace=True)
df['Gross'] = df['Gross'] / 1000000 # CAUSE OF BIG VALUES WE`LL USE GROSS COUNTING IN $mln
```

```python
df.isna().sum()
```

**START ANALYSIS//**

```python
plt.subplot(1, 2, 1)
sns.histplot(data = df['Movie Rating'], color="#c10000", kde=True)
plt.subplot(1, 2, 2)
sns.histplot(data = df['Gross'], kde=True)
```

```python
sns.scatterplot(data=df, x="Gross", y="Movie Rating")
```

```python
# Sort films
gross_counts = df.nlargest(10, 'Gross')[['Movie Name', 'Gross']]

# Graphic
plt.figure(figsize=(12, 8))
sns.barplot(x='Movie Name', y='Gross', data=gross_counts.head(10), palette='viridis')

# Add labels and title
plt.xlabel('Movie Name')
plt.ylabel('Gross')
plt.title('Top 10 Highest-Grossing Films')
plt.xticks(rotation=45)

# Show plot
plt.grid(True)
plt.show()
```
